**HANDLE OUTLIERS**

```python
import numpy as np
import pandas as pd

# Function to remove outliers for specific numerical columns using percentile-based filtering
def remove_outliers_percentile(dataframe, columns):
    new_df = dataframe.copy()
    
    for column in columns:
        if column in dataframe.select_dtypes(include=['number']).columns:  # Ensure column is numeric
            temp_df = new_df.sort_values(by=column)  # Sort values in ascending order
            temp_df.loc[:, 'index'] = range(1, len(temp_df) + 1)  # Create an index column
            
            low = round(len(temp_df) * 0.01)  # 1% lower bound
            upp = len(temp_df) - round(len(temp_df) * 0.04)  # 4% upper bound
            
            # Create an outlier flag (1 = keep, 0 = remove)
            temp_df['outl'] = np.where((temp_df['index'] > low) & (temp_df['index'] <= upp), 1, 0)
            
            # Keep only non-outliers
            temp_df = temp_df[temp_df['outl'] == 1].drop(columns=['index', 'outl'])  # Drop helper columns
            new_df = temp_df
        else:
            print(f"Warning: Column '{column}' is not numeric and was skipped.")
    
    return new_df

# Specify columns to apply outlier removal
columns_to_clean = ['beans_kgs_hh_seed', 'irish_potatoes_bags_hh_seed','Coffee_total_yield']  # Replace with actual column names

# Apply function to dataset
IDS_8 = remove_outliers_percentile(IDS_7, columns_to_clean)

# Display shape before and after
print(f"Original shape: {IDS_7.shape}")
print(f"Cleaned shape: {IDS_8.shape}")

# # Preview cleaned data
# print(cleaned_IDS_4.head())

# Optional: Save cleaned dataset
# cleaned_IDS_4.to_csv("cleaned_IDS_4.csv", index=False)

```

**PRINT OUTLIERS**

```python
# Function to count outliers per column using IQR
def count_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return ((data < lower_bound) | (data > upper_bound)).sum()

# Apply to numerical columns
outlier_counts = IDS_9.select_dtypes(include=['number']).apply(count_outliers_iqr)
print("Outlier counts per column:\n", outlier_counts)
```

*FUNCTION TO REMOVE OUTLIERS*

```python

# Define a function to remove outliers (as before)
def remove_outliers_percentile(dataframe, columns):
    new_df = dataframe.copy()
    
    for column in columns:
        if column in dataframe.select_dtypes(include=['number']).columns:
            temp_df = new_df.sort_values(by=column)
            temp_df.loc[:, 'index'] = range(1, len(temp_df) + 1)
            
            low = round(len(temp_df) * 0.01)
            upp = len(temp_df) - round(len(temp_df) * 0.04)
            
            temp_df['outl'] = np.where((temp_df['index'] > low) & (temp_df['index'] <= upp), 1, 0)
            temp_df = temp_df[temp_df['outl'] == 1].drop(columns=['index', 'outl'])
            new_df = temp_df
        else:
            print(f"Warning: Column '{column}' is not numeric and was skipped.")
    
    return new_df
```

*Impuitng Misssing Values*

```python
# Define a function for imputing missing values
def impute_missing_values(df, column_name):
    knn_imputer = KNNImputer(n_neighbors=4)
    df[column_name] = knn_imputer.fit_transform(df[[column_name]])
    return df
```

*Preprocessor Pipieline*

```python
# Define the preprocessor pipeline
def build_preprocessor():
    # Define transformers for each step of preprocessing
    numeric_transformer = Pipeline(steps=[
        ('imputer', FunctionTransformer(lambda x: impute_missing_values(x, 'Agriculture Value (USD)'))),  # Imputation
        ('outlier_remover', FunctionTransformer(lambda x: remove_outliers_percentile(x, ['beans_kgs_hh_seed', 'irish_potatoes_bags_hh_seed', 'Coffee_total_yield']))),  # Outlier Removal
        ('scaler', MinMaxScaler())  # Scaling
    ])
    
    # ColumnTransformer to apply transformations to numeric columns
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, ['beans_kgs_hh_seed', 'Coffee_total_yield'])  # Apply to relevant numeric columns
        ],
        remainder='passthrough'  # Leave other columns unchanged (optional)
    )
    
    return preprocessor
```

*SGD Pipeline*

```python
# Define the pipeline with preprocessor and model
def build_pipeline():
    # Define the model (SGD Regressor in this case)
    model = SGDRegressor(penalty='l2', max_iter=10000000, random_state=42)
    
    # Create the full pipeline
    pipeline = Pipeline(steps=[
        ('preprocessor', build_preprocessor()),  # Apply preprocessing
        ('model', model)  # Train the model
    ])
    
    return pipeline
```

**USING A CUSTOM CLASS**

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd

# Define the custom class for handling scaling and prediction
class ScalableModel:
    def __init__(self, model, feature_scaler=MinMaxScaler(), target_scaler=MinMaxScaler()):
        self.model = model
        self.feature_scaler = feature_scaler
        self.target_scaler = target_scaler
        
    def fit(self, X_train, y_train):
        # Fit the feature scaler and transform the features
        self.X_train_scaled = self.feature_scaler.fit_transform(X_train)
        
        # Fit the target scaler and transform the target variable
        self.y_train_scaled = self.target_scaler.fit_transform(y_train.values.reshape(-1, 1))
        
        # Train the model on scaled data
        self.model.fit(self.X_train_scaled, self.y_train_scaled)
        
    def predict(self, X_test):
        # Scale the test features
        X_test_scaled = self.feature_scaler.transform(X_test)
        
        # Get the scaled predictions from the model
        predictions_scaled = self.model.predict(X_test_scaled)
        
        # Inverse transform the predictions to get them in the original scale
        predictions_raw = self.target_scaler.inverse_transform(predictions_scaled.reshape(-1, 1))
        
        return predictions_raw.flatten() 
```
